```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merge_df = pd.merge(study_results, mouse_metadata, on="Mouse ID", how='outer')

# Display the data table for preview
merge_df
```

```python
#Number of unique mice
merge_df["Mouse ID"].nunique()
```

```python
#sort data by timepoint
sorted_df=merge_df.sort_values(['Timepoint','Mouse ID']).reset_index()
sorted_df=sorted_df.drop(['index'], axis=1)
#Duplicate mouse id data
duplicate_df=sorted_df[sorted_df.duplicated(['Mouse ID','Timepoint'],keep=False)]
duplicate_df
```

```python
#delete Mouse g989
cleaned_df=sorted_df.drop(sorted_df[sorted_df['Mouse ID']=="g989"].index)
cleaned_df.head()
```

```python
#Check for new number of mice
cleaned_df["Mouse ID"].nunique()
```

```python
#group by Drug Regimen
grouped_df=cleaned_df.groupby(cleaned_df['Drug Regimen'])
#pandas descibe
regimen_info1=grouped_df['Tumor Volume (mm3)'].describe()
#median
#median=grouped_df['Tumor Volume (mm3)'].median()
#variance
var = grouped_df['Tumor Volume (mm3)'].var(ddof=1)
#standard error of the mean
sem=grouped_df['Tumor Volume (mm3)'].sem()

#combined table
regimen_info=pd.merge(regimen_info1, var, on="Drug Regimen")
regimen_info=pd.merge(regimen_info, sem, on="Drug Regimen")
#delete columns
regimen_info = regimen_info.drop(columns=['count','min','25%','75%','max'])
#rename
regimen_info=regimen_info.rename(columns={"50%":"median",
                                          "Tumor Volume (mm3)_x":"var",
                                          "Tumor Volume (mm3)_y":"sem"})
#reorder
regimen_info=regimen_info[["mean","median","var","std","sem"]]
regimen_info
```
